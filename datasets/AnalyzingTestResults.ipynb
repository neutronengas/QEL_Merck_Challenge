```python
import pandas as pd
```

```python
results = pd.read_excel("TestResults.xlsx")
results.columns
```

```python
results["Quantum Proportion"] = results["Quantum Solution"] / results["Classical Optimal Solution"]
results["Classical Proportion"] = results["Classical Greedy Solution"] / results["Classical Optimal Solution"]
```

```python
plotting_df = results.groupby("Problem type")[["Quantum Proportion", "Classical Proportion"]].mean().transpose()[["alm_str", "weak_cor", "strong_cor", "subset"]]
plotting_df.columns = ["Almost strong correlated", "Weakly correlated", "Strongly correlated", "Subset"]
plotting_df = plotting_df.transpose()
plot = plotting_df.plot(kind="bar", color=["#558F3B", "#000000"])
plot.legend(bbox_to_anchor=(1.05, 1.0))
plot.set_ylabel("Average obtained proportion of optimal solution")
plot.set_xlabel("Problem type")
plot.set_ylim([0.8, 1])
```
